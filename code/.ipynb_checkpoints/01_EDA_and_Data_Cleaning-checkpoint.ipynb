#### Saving T2 Data

```python
# creating a list of dictionaries for control group
control_arrays_dict = array_dict('control', 't2', 'all', 'dcm')
```

```python
# creating a dataframe from list of dictionaries for control group
control_df = pd.DataFrame(control_arrays_dict, columns=['path', 'scan'])

# patients identification
control_df['patient'] = [each[46:55] for each in control_df['path']]

# study identification
control_df['study'] = [each[34:45] for each in control_df['path']]

# MRI scan type
control_df['scan_type'] = [each[66:80] for each in control_df['path']]

# index of slice
control_df['slice'] = [each[100:103] for each in control_df['path']]

# schizophrenia diagnosis
control_df['schizophrenia'] = 0
```

```python
# creating a list of dictionaries for schiz group
schiz_arrays_dict = array_dict('schiz', 't2', 'all', 'dcm')
```

```python
# creating dataframe from list of dicionaries for schiz group
schiz_df = pd.DataFrame(schiz_arrays_dict, columns=['path', 'scan'])

# patient identification
schiz_df['patient'] = [each[42:51] for each in schiz_df['path']]

# study identification
schiz_df['study'] = [each[30:41] for each in schiz_df['path']]

# MRI scan type
schiz_df['scan_type'] = [each[62:76] for each in schiz_df['path']]

# index of slice
schiz_df['slice'] = [each[96:99] for each in schiz_df['path']]

# schizophrenia diagnosis
schiz_df['schizophrenia'] = 1
```

```python
# concating dataframes
t2_df = pd.concat([schiz_df, control_df], axis=0)
```

```python
# dropping path column
t2_df = t2_df.drop(columns='path')
```

```python
# saving t2 dataframe to csv
# saves a lot of space
t2_df.to_csv('t2_scans.csv')
```

### DCM to PNG (prep for gif)

**Note**: png folder deleted upon upload to ensure no leakage of data when uploaded

```python
path = '../brain_images/no_known_disorder/cobre_07325/A00000300/Session_2/t2_tse_tra_192_0005/'
path_2 = '../brain_images/no_known_disorder/cobre_07325/A00000300/Session_2/mprage_5e_RMS_0003/'
```

```python
# png of slices for each folder in path & path_2
for each in os.listdir(path):
    if 'dcm' in each:
        im = pydicom.dcmread(path+each)
        im = im.pixel_array.astype(float)
        scaled_image = (np.maximum(im, 0) / im.max()) * 255.0
        scaled_image = np.uint8(scaled_image)
        final_image = Image.fromarray(scaled_image)
        final_image.save('../png/t2/'+each+'.png')

for each in os.listdir(path_2):
    if 'dcm' in each:
        im = pydicom.dcmread(path_2+each)
        im = im.pixel_array.astype(float)
        scaled_image = (np.maximum(im, 0) / im.max()) * 255.0
        scaled_image = np.uint8(scaled_image)
        final_image = Image.fromarray(scaled_image)
        final_image.save('../png/mprage/'+each+'.png')
```

#### Dcm to Nii with library

**Note**: WinError occurs but nii files are still made

```python
from nipype.interfaces.dcm2nii import Dcm2niix
```

```python
# setting initial control scan path
control_dir = '../brain_images/no_known_disorder/cobre_07325/'

# setting total control patients
patients_c = [each for each in os.listdir(control_dir)]
```

```python
# converting to nii file type for each dir for all control images
for patient in patients_c:
    for session in os.listdir(control_dir+patient):
        for scan in os.listdir(control_dir+patient+'/'+session): 
            if "t2_tse_tra_192_0" in scan:
                try:
                    converter = Dcm2niix()
                    converter.inputs.source_dir = control_dir+patient+'/'+session+'/'+scan+'/'
                    converter.inputs.compress = 'n'
                    converter.inputs.output_dir = control_dir+patient+'/'+session+'/'+scan+'/'
                    converter.run()
                except:
                    print(f'Error for file: {each}')
```

```python
# setting initial schizophrenic image path
schiz_dir = '../brain_images/schizophrenic/cobre_07325/'

# setting total schizophrenic patients
patients_s = [each for each in os.listdir(schiz_dir)]
```

```python
# converting to nii file type for each dir for all schizophrenic images
for patient in patients_s:
    for session in os.listdir(schiz_dir+patient):
        for scan in os.listdir(schiz_dir+patient+'/'+session): 
            if "t2_tse_tra_192_0" in scan:
                try:
                    converter = Dcm2niix()
                    converter.inputs.source_dir = schiz_dir+patient+'/'+session+'/'+scan+'/'
                    converter.inputs.compress = 'n'
                    converter.inputs.output_dir = schiz_dir+patient+'/'+session+'/'+scan+'/'
                    converter.run()
                except:
                    print(f'Error for file: {each}')
```

```python
# deleting unnecessary files -- be careful!
for patient in patients_c:
    for session in os.listdir(control_dir+patient):
        for scan in os.listdir(control_dir+patient+'/'+session): 
            if "t2_tse_tra_192_0" in scan:
                for each in os.listdir(control_dir+patient+'/'+session+'/'+scan+'/'):
                    if 'nii' in each:
                        # print(each)
                        os.remove(control_dir+patient+'/'+session+'/'+scan+'/'+each)   
```
